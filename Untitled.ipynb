```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib.animation
import matplotlib.lines as lines
```

**Question 1**: Write code which takes $(e, l_z, r_0)$ as input and returns $t(\tau)$, $\phi(\tau)$, $r(\tau)$, and $u_r(\tau)$ as output for $\tau > 0$. 

$$\frac{dt}{d\tau} = e \left( 1 - \frac{2GM}{rc^2} \right)^{-1}$$
$$\frac{d\phi}{d\tau} = \frac{l_z}{r^2}$$
$$\frac{dr}{d\tau} = u^r$$
$$\frac{du^r}{d\tau} = - \frac{GM}{r^2} + \frac{l_z^2}{r^3} - \frac{3GMl_z^2}{r^4c^2} $$

```python
# scipy's ODE integrator has a cool capability where it can take a vector as 
# an input, then use that to return the vectorized version of df/dt. We must 
# therefore define the differential equation as a single vector. 

def schwartzchild_coord_velocities(y, tau):
    [t, phi, r, ur, e, lz] = y
    dt_dtau = e*(1-(2/r))**-1
    dphi_dtau = lz/r**2
    dr_dtau = ur
    dur_dtau = -1/r**2 + lz**2/r**3 - (3*lz**2)/(r**4)
    return [dt_dtau, dphi_dtau, dr_dtau, dur_dtau, 0, 0]
```

$$u^r = \pm \sqrt{\frac{e^2}{c^2} - \left(1 - \frac{2GM}{r_0c^2} \right) \left(\frac{l_z^2}{r_0^2} + c^2\right)}$$

```python
def ur0(e, lz, r0): return np.sqrt(e**2 - (1 - 2/r0)*(lz**2/r0**2 + 1))

e = 0.98
lz = 4
r0 = 12
ur = ur0(e, lz, r0)
tau = np.linspace(0, 10000, 10000)
results = odeint(schwartzchild_coord_velocities, [0, 0, r0, ur, e, lz], tau)
```

```python
r = results[:,2]
phi = results[:,1]
x = r*np.cos(phi)
y = r*np.sin(phi)
fig, ax = plt.subplots(figsize=(8,8))
plt.axis('equal')
ax.plot(x, y)
```
